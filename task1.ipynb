```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
# now importing the dataset
data = pd.read_csv('http://bit.ly/w-data')
print(data.shape)
data.head(10)
```

```python
# plotting our dataset
plt.scatter(data = data , x = "Hours", y="Scores")
plt.xlabel('number of study hours')
plt.ylabel('scores obtained')
plt.show
```

This scatter plot tells that the relation between the score and study hours is simple linear. Or we can also plot the regression line

```python
# plotting the regression line
sns.regplot(data = data , x = 
"Hours", y="Scores")
plt.xlabel('number of study hours')
plt.ylabel('scores obtained')
plt.show
```

So this is a simple linear regression

```python
# Train test split
data_train, data_test = train_test_split(data,test_size=0.2, random_state=0)
```

```python
print(data_train.shape)
print(data_test.shape)
```

```python
y_train = data_train.pop('Scores')

x_train = data_train

y_test = data_test.pop('Scores')
x_test = data_test
```

```python
#let's train our model
lr = LinearRegression() 
data_reg = lr.fit(x_train, y_train)
```

```python
# verify the r2 score for trained data
data_reg.score(x_train, y_train)
```

```python
# verify the r2 value for test data
data_reg.score(x_test, y_test)
```

```python
print(x_test)
y_pred = lr.predict(x_test)
print(y_pred)
```

```python
# compare actual and predicted score 
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```python
# What will be predicted score if a student studies for 9.25 hrs/ day?
time = np.array([9.25]).reshape(-1,1)
new_pred = lr.predict(time)
print("Predicted Score = {}".format(new_pred[0]))
```
